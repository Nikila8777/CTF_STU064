```python
import os
import json
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.metrics import silhouette_score

import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

# Colab helpers (try to import; will be None outside Colab)
try:
    from google.colab import files, drive
except Exception:
    files = None
    drive = None
```

```python
print("Shape:", df.shape)
print("\nColumns & dtypes:\n", df.dtypes)
print("\nMissing values per column:\n", df.isna().sum())

# Simple guidance for columns with many missing values
thresh = 0.5 * df.shape[0]
drops = [c for c in df.columns if df[c].isna().sum() > thresh]
if drops:
    print("\nColumns with >50% missing (consider dropping):", drops)
else:
    print("\nNo columns with >50% missing values.")

# Show basic stats for numeric columns
display(df.describe(include=[np.number]).round(3))
```

```python
numeric_features = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = df.select_dtypes(include=['object','category','bool']).columns.tolist()

print("Numeric features:", numeric_features)
print("Categorical features:", categorical_features)

# If you need to drop ID/text columns, put them here:
drop_cols = []  # e.g. ['customer_id']
drop_cols = [c for c in drop_cols if c in df.columns]
if drop_cols:
    df = df.drop(columns=drop_cols)
    print("Dropped columns:", drop_cols)
```

```python
# Numeric pipeline: median impute + standard scaling
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Categorical pipeline: most frequent impute + onehot (dense matrix)
# NOTE: use sparse_output=False for modern scikit-learn
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
], remainder='drop')

# Fit-transform to get numeric matrix X
X = preprocessor.fit_transform(df)

# Compose feature names for postprocessing / centroids
feature_names = list(numeric_features)
if categorical_features:
    # get_feature_names_out available in modern sklearn
    ohe = preprocessor.named_transformers_['cat'].named_steps['onehot']
    try:
        cat_ohe_names = ohe.get_feature_names_out(categorical_features).tolist()
    except Exception:
        # fallback (older versions): create simple names
        cat_ohe_names = []
        for c in categorical_features:
            vals = sorted(df[c].dropna().unique().tolist())
            cat_ohe_names += [f"{c}__{v}" for v in vals]
    feature_names += cat_ohe_names

print("Transformed feature matrix shape:", X.shape)
print("Number of features after preprocessing:", len(feature_names))
```

```python
n_comp = min(10, X.shape[1])
pca = PCA(n_components=n_comp, random_state=42)
X_pca = pca.fit_transform(X)
explained = np.cumsum(pca.explained_variance_ratio_)
print(f"Explained variance by first {n_comp} PCA components:\n", np.round(explained, 4))

# 2D and 3D arrays for plotting
X_2d = X_pca[:, :2]
X_3d = X_pca[:, :3] if X_pca.shape[1] >= 3 else None
```

```python
def choose_k(X, kmin=2, kmax=12):
    inertias = []
    silhouettes = []
    K_range = list(range(kmin, kmax+1))
    for k in K_range:
        km = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = km.fit_predict(X)
        inertias.append(km.inertia_)
        if len(set(labels)) > 1:
            silhouettes.append(silhouette_score(X, labels))
        else:
            silhouettes.append(np.nan)
    return K_range, inertias, silhouettes

K_range, inertias, silhouettes = choose_k(X, 2, 12)

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(K_range, inertias, '-o')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow: Inertia vs k')

plt.subplot(1,2,2)
plt.plot(K_range, silhouettes, '-o')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs k')

plt.tight_layout()
plt.show()
```

```python
n_clusters = int(input("Enter number of clusters to use (suggest from elbow/silhouette). Default 4: ") or 4)
# Use KMeans (or MiniBatchKMeans for very large datasets)
km = KMeans(n_clusters=n_clusters, random_state=42, n_init=20)
labels = km.fit_predict(X)
df['cluster'] = labels
print("Cluster distribution:")
print(df['cluster'].value_counts().sort_index())
```

```python
centroids = km.cluster_centers_
cent_df = pd.DataFrame(centroids, columns=feature_names)
cent_df.index.name = 'cluster'
display(cent_df.round(4))

# Note: centroids are in preprocessed feature space (scaled numeric & one-hot encoded cat columns).
# If you want numeric centroids approximately in original units, you can invert the scaler for numeric parts.
```

```python
# Numeric summaries per cluster
profile_num = df.groupby('cluster')[numeric_features].agg(['mean','median','count']).round(3)
profile_num.columns = ['_'.join(col).strip() for col in profile_num.columns.values]
profile_num = profile_num.reset_index()
display(profile_num)

# Categorical modes per cluster
if categorical_features:
    cat_profile = {}
    for c in categorical_features:
        cat_profile[c] = df.groupby('cluster')[c].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
    cat_profile_df = pd.DataFrame(cat_profile).reset_index()
    display(cat_profile_df)
```
